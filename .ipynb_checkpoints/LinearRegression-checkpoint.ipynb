```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate the dataset 

np.random.seed(42) # for reproducability

m = 3.5 # slope
b = 1.5 #intercept
X = np.random.uniform(0, 20, 200) # datapoints
noise = np.random.normal(0, 10, len(X))
Y = m * X + b + noise # y = mx + b + alpha 
plt.scatter(X, Y, label="Data points")
plt.plot(X, m * X + b, color="red", label="True Line (no noise)")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Generated dataset")
plt.legend()
plt.show()
plt.savefig("output_graph.png")
```

## Initialize Values and define functions to make prediction, measure MSE

```python
class MyLinearRegression:
    def __init__(self, alpha, max_iter, threshold):
        """
        A linear regression model that takes in a dataset and learns a line that minimizes MSE.
        Args:
            alpha - the learning rate
            max_iter - the # iterations during training
        """
        self._alpha = alpha
        self._max_iter = max_iter
        self._threshold = threshold
        self._X = None
        self._Y = None
        self._m = None
        self._b = None

    def get_m(self):
        return self._m
    def get_b(self):
        return self._b
    
    def fit(self, X, Y):
        """Fits the model on the data and optimizes values for m, b. Outputs final MSE.
        Args:
            X - the feature which we are using to predict Y
            Y - the feature we are predicting
        """
        self._X = X
        self._Y = Y
        self._m, self._b = 0, 0
        mse = self.mse()
        print(f"Initial m = 0, Initial b = 0, Initial MSE = {mse}")
        print("Beginning training...")
        self.gradient_descent(self._alpha, self._max_iter, self._threshold)

    def predict(self, x):
        """
        Makes a single prediction based on a given value for x.
        Args:
            x - the input with which to predict
        Returns:
            y_pred - the predicted output
        """
        return self._m * x + self._b

    def make_predictions(self) -> np.ndarray:
        """
        Make predictions based on the current slope and bias
        Returns:
            Y_pred - the predicted y values based on current m and b 
        """
        Y_pred = self._X * self._m + self._b
        return Y_pred

    def mse(self) -> int:
        """
        Our loss function. Gets the mean squared error of the current line and the points. (the distance between points and line)
        Returns:
            mse (int) - the mean squared error 
        """
        squared_errors = (self._Y - self.make_predictions()) ** 2 # y and y_pred are numpy array with same shape, this gets squared residuals
        return np.mean(squared_errors)
        
    def gradient(self) -> tuple[float, float]:
        """Computes the gradient of dm and db (the gradient for m and gradient for b). Proof will be attached as a pdf to the repo."""
        n = len(self._X)
        y_hat = self.make_predictions()
        error = y_hat - self._Y # residuals
        dm = (2 / n) * np.sum(error * self._X)
        db = (2 / n) * np.sum(error)
        return dm, db
        
        
    def gradient_descent(self, alpha, max_iter, threshold): 
        """ 
        Performs gradient descent until the mean squared error dips below the threshold.
        Args:
            alpha (learning rate) - what to scale the gradient by on each epoch
            max_iter - maximium # of iterations before stop 
        """
        i = 0
        
        while i <= max_iter:
            dm, db = self.gradient()
            self._m -= alpha * dm
            self._b -= alpha * db
            mse = self.mse()
            if mse < threshold:
                print(f"Stopping at iteration {i}.")
                break
            if i % 100 == 0: 
                print(f"Iteration {i}: MSE = {self.mse():.4f}")
            i += 1
        
```

## Using the model


```python
model = MyLinearRegression(alpha=0.0075, max_iter=200, threshold=15)
model.fit(X, Y)
m, b = model.get_m(), model.get_b()
print("m: ", m, "b: ", b)
```

## Plotting results and comparing with model from sklearn

```python
from sklearn.linear_model import LinearRegression
np.random.seed(42)
model2 = LinearRegression()
X_reshaped = X.reshape(-1, 1)
model2.fit(X_reshaped, Y)
m_sklearn, b_sklearn = model2.coef_, model2.intercept_
print("m: ", m_sklearn[0], "b: ", b_sklearn)
```

```python
plt.scatter(X, Y, label="Data points", color="blue", alpha=0.6)

# Plot the sklearn line
plt.plot(X, m_sklearn * X + b_sklearn, color="orange", label="Sklearn Line")
plt.plot(X, 3.5 * X + 1.5, color="red", label="Real Line")
# Plot your custom model line
plt.plot(X, m * X + b, color="green", linestyle="--", label="MyLinearRegression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Comparison of Sklearn and MyLinearRegression")
plt.legend()
plt.show()
plt.savefig("output_graph.png")
```

```python
# compare MSE from our model vs sklearn model
# define mse again for clarity

def mse(X, Y, Y_pred):
    """Calculates mean squared error between predicted and actual values."""
    n = len(Y)
    error = Y_pred - Y
    return (1 / n) * np.sum(error**2)
sk_pred = X * m_sklearn + b_sklearn
my_pred = X * m + b
real_line = X * 3.5 + 1.5
print(f"Sklearn model MSE: {mse(X, Y, sk_pred)}")
print(f"My model MSE: {mse(X, Y, my_pred)}")
print(f"Real Line: {mse(X, Y, real_line)}")
```
